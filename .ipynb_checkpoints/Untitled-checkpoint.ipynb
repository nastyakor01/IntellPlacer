```python
from skimage.feature import canny, corner_harris, corner_peaks, corner_fast, corner_subpix, match_descriptors, ORB
from scipy.ndimage.morphology import binary_closing
from skimage.transform import warp, AffineTransform
import cv2
from imageio import imread
from skimage.color import rgb2gray
import os
import csv

import matplotlib.pyplot as plt
import matplotlib.patches
from matplotlib import cm
import numpy as np
```

```python
treshold = 0.7
```

```python
def preprocess(image_path):
    img = imread(image_path)

    img_gray = rgb2gray(img)
    img_edges = canny(img_gray, sigma=1, low_threshold=0.01)

    return binary_closing(img_edges, iterations = 3)


def find_keypoints(edge_map, r):
    keypoints = corner_peaks(corner_harris(edge_map), threshold_rel=r)
    return keypoints
```

```python
# preprocess templates
images = os.listdir('objects')
rectangles = {image: None for image in images}
keypoints = {image: None for image in images}
for image in images:
    points = find_keypoints(preprocess(f'objects/{image}'), treshold)
    keypoints[image] = points
    rectangles[image] = cv2.minAreaRect(points)
```
